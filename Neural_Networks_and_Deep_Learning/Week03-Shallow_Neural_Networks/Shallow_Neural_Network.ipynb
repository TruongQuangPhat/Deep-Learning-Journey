## Neural Networks Overview
### 1. Evolution from Logistic Regression to Neural Networks
In Logistic Regression, the computation graph consists of a single processing block (node) that performs two steps:
- **Step 1:** Compute the linear value $z = w^T x + b$.
- **Step 2:** Compute the activation value $a = \sigma(z)$, which simultaneously serves as the output prediction $\hat{y}$.

  Subsequently, the loss function $\mathcal{L}$ is computed based on $a$ and the actual label $y$.

In an artificial neural network, the model does not merely have one node but comprises multiple layers, each containing multiple interconnected nodes (neurons). The process of computing the paired linear value $z$ and activation value $a$ is sequentially repeated across each layer before producing the final prediction.


### 2. New Mathematical Notation Conventions
To represent the multi-layer structure of a neural network, a new superscript notation system is established to distinguish parameters and variables between layers:
- **Square brackets $[l]$:** Denote the layer index within the neural network. For example, $z^{[1]}$ and $a^{[1]}$ are the values computed at the first layer; $W^{[2]}$ and $b^{[2]}$ are the weight matrix and bias vector of the second layer.
- **Parentheses $(i)$:** It is necessary to clearly distinguish layer brackets from parentheses. Parentheses are used to denote the index of a training example. For example, $x^{(i)}$ is the $i$-th input data sample in the dataset.

### 3. Computation Flow in Neural Networks
Considering a simple 2-layer neural network, the operational process occurs in two main phases:

**Forward Propagation:**

The computation proceeds from left to right, passing the data flow through each layer to generate a prediction:
- **At layer 1:** Starting from the input $x$, the linear value $z^{[1]}$ is computed using the parameters $W^{[1]}, b^{[1]}$, and then passed through an activation function to obtain $a^{[1]}$.
- **At layer 2:** Taking $a^{[1]}$ as the new input data, the linear value $z^{[2]}$ is computed using the parameters $W^{[2]}, b^{[2]}$, and then passed through an activation function to obtain $a^{[2]}$. In this 2-layer network, $a^{[2]}$ acts as the final output prediction $\hat{y}$.
- Finally, the loss function $\mathcal{L}$ is computed based on the discrepancy between $a^{[2]}$ and the actual label $y$.

**Backward Propagation:**

The derivative computation proceeds in the reverse direction from right to left, starting from the loss function to update the parameters for the entire network:
- The algorithm begins by computing the derivatives at the final layer (layer 2): $da^{[2]}$, $dz^{[2]}$, from which the parameter derivatives of layer 2, $dW^{[2]}$ and $db^{[2]}$, are deduced.
- This error quantity is continually propagated backward to the preceding layer to compute the corresponding variations $da^{[1]}$ and $dz^{[1]}$, thereby allowing the computation of the layer 1 parameter derivatives, $dW^{[1]}$ and $db^{[1]}$.

## Neural Network Representation
### 1. Structure of the Neural Network Layers
A standard neural network is divided into three main regions:
- **Input Layer:**
  - Contains the input data features, for example, $x_1, x_2, x_3$.
  - The variable $a$ is used as an abbreviation for activation value. The input layer directly passes the data $x$ into the network, thus establishing the convention $a^{[0]} = x$.
- **Hidden Layer:**
  - The term hidden stems from the characteristics of the training dataset. The training set only provides the input values $x$ and the target labels $y$; therefore, the computational processes and actual values generated at these intermediate nodes are completely unobserved from the outside.
  - Assuming the hidden layer has 4 nodes (hidden units), the output activation values of this layer are grouped into a 4-dimensional column vector, denoted as $a^{[1]}$. The elements within this vector include $a_1^{[1]}, a_2^{[1]}, a_3^{[1]}$, and $a_4^{[1]}$.
- **Output Layer:**
  - In a basic problem, this layer typically contains a single node responsible for returning the final predicted value.
  - The activation value of this layer is denoted as $a^{[2]}$, and this is precisely the prediction $\hat{y}$ (meaning $\hat{y} = a^{[2]}$).

### 2. Layer Counting Convention
According to international machine learning conventions, when counting the total number of layers in a neural network, the input layer (layer 0) is not included.

Therefore, an architecture comprising 1 input layer, 1 hidden layer, and 1 output layer is officially referred to as a **2-layer neural network**.

### 3. Parameter Matrix Dimensions
Each computational layer (layer 1 and layer 2) is associated with a distinct set of parameters consisting of a weight matrix $W$ and a bias vector $b$. Based on the number of nodes in each layer, the dimensions of these matrices are strictly determined to ensure valid linear algebra multiplication:
- Layer 1 parameters ($W^{[1]}$ and $b^{[1]}$): The input layer has 3 features and the first hidden layer has 4 nodes. Consequently, the weight matrix $W^{[1]}$ has the dimension $(4, 3)$ and the bias vector $b^{[1]}$ has the dimension $(4, 1)$.
- Layer 2 parameters ($W^{[2]}$ and $b^{[2]}$): The first hidden layer has 4 nodes transmitting data to the output layer which has only 1 node. Consequently, the weight matrix $W^{[2]}$ has the dimension $(1, 4)$ and the bias vector $b^{[2]}$ has the dimension $(1, 1)$.

## Computing a Neural Network's Output
### 1. Computational Breakdown for a Single Neuron
In Logistic Regression, a predictive model involves two computational steps: calculating the linear value $z$ and calculating the activation value $a$.

A neural network applies this exact process to each of its internal neurons. Considering the first neuron of the first hidden layer, the two computational steps are written as:
$$
z_1^{[1]} = (w_1^{[1]})^T \cdot x + b_1^{[1]}
$$
$$
a_1^{[1]} = \sigma(z_1^{[1]})
$$

Where:
- The superscript $[1]$ denotes variables and parameters belonging to the first layer.
- The subscript $1$ denotes the position of the neuron (neuron number 1) within that layer.
- Neurons 2, 3, and 4 in the same hidden layer independently perform a similar set of equations to produce $a_2^{[1]}, a_3^{[1]}$, and $a_4^{[1]}$.

<img src="../images/Week03/image3.png" width=600>

### 2. Vectorizing the Hidden Layer (Layer 1)
Using a for loop to sequentially compute values for the 4 neurons would make the program run very slowly. Therefore, vectorization is required by grouping the parameters into matrices.

**Parameter matrix setup:**
- By vertically stacking the 4 row vectors $(w_1^{[1]})^T$ through $(w_4^{[1]})^T$, we create a weight matrix $W^{[1]}$ of dimension $(4, 3)$. (Where $4$ is the number of hidden layer neurons, and $3$ is the number of input features).
- By vertically stacking the 4 bias terms $b_1^{[1]}$ through $b_4^{[1]}$, we create a column vector $b^{[1]}$ of dimension $(4, 1)$.

**Vectorized equations for layer 1:**

The entire computational process of the 4 hidden layer neurons is now consolidated into two linear algebra equations:
$$
Z^{[1]} = W^{[1]} \cdot x + b^{[1]}
$$
$$
A^{[1]} = \sigma(Z^{[1]})
$$
The returned result is a column vector $A^{[1]}$ of dimension $(4, 1)$ containing the activation values of the entire hidden layer.

### 3. Computing the Output Layer (Layer 2)
The output layer operates exactly like an independent Logistic Regression model, but instead of receiving raw data $x$, it receives the processed data $A^{[1]}$ passed from the hidden layer.
- The parameters for this layer are the matrix $W^{[2]}$ of dimension $(1, 4)$ and the scalar $b^{[2]}$ of dimension $(1, 1)$.
- The forward propagation equations at layer 2 are written as:
  $$
  Z^{[2]} = W^{[2]} \cdot A^{[1]} + b^{[2]}
  $$
  $$
  A^{[2]} = \sigma(Z^{[2]})
  $$
  The obtained value $A^{[2]}$ is a real number, representing the final prediction $\hat{y}$ of the entire network.

<img src="../images/Week03/image4.png" width=600>

### 4. Summary
To compute the output for a raw data sample $x$ (which is uniformly denoted as $A^{[0]}$), you only need to implement these exact 4 mathematical equations in Python:
- **Equation 1:** $$Z^{[1]} = W^{[1]} \cdot A^{[0]} + b^{[1]}$$
- **Equation 2:** $$A^{[1]} = \sigma(Z^{[1]})$$
- **Equation 3:** $$Z^{[2]} = W^{[2]} \cdot A^{[1]} + b^{[2]}$$
- **Equation 4:** $$A^{[2]} = \sigma(Z^{[2]})$$

## Vectorizing Across Multiple Examples
### 1. Unvectorized Method (Using Loops)
To compute predictions for all $m$ training examples without utilizing vectorization, one must establish a loop running from $i = 1$ to $m$. At each iteration, the neural network executes 4 forward propagation equations for a single data sample $x^{(i)}$:
$$
z^{[1](i)} = W^{[1]}x^{(i)} + b^{[1]}
$$
$$
a^{[1](i)} = \sigma(z^{[1](i)})
$$
$$
z^{[2](i)} = W^{[2]}a^{[1](i)} + b^{[2]}
$$
$$
a^{[2](i)} = \sigma(z^{[2](i)})
$$
**Notation conventions:**
- **Square bracket superscript $[l]$:** Represents the layer index in the neural network (e.g., layer 1, layer 2).
- **Parentheses superscript $(i)$:** Represents the index of the data sample in the training set (e.g., the $i$-th sample).

### 2. Vectorized Method (Using Matrices)
Instead of processing each column vector $x^{(i)}$ discretely, the vectorization technique aggregates all $m$ input vectors into a single matrix by stacking them horizontally side-by-side.

The input matrix $X$ with dimensions $(n_x, m)$ is defined as follows:
$$
X = 
\begin{bmatrix} 
| & | & & | \\ 
x^{(1)} & x^{(2)} & \dots & x^{(m)} \\ 
| & | & & | 
\end{bmatrix}
$$
Similarly, the resulting vectors $z$ and $a$ generated from each sample are also stacked horizontally to form the uppercase matrices $Z$ and $A$:
$$
Z^{[1]} = \begin{bmatrix} | & | & & | \\ z^{[1](1)} & z^{[1](2)} & \dots & z^{[1](m)} \\ | & | & & | \end{bmatrix}
$$
$$
A^{[1]} = \begin{bmatrix} | & | & & | \\ a^{[1](1)} & a^{[1](2)} & \dots & a^{[1](m)} \\ | & | & & | \end{bmatrix}
$$
Thanks to this matrix arrangement, the for loop is completely eliminated. The forward propagation process for all $m$ samples is now executed using just 4 highly concise matrix equations:
- Equation 1: $$Z^{[1]} = W^{[1]}X + b^{[1]}$$ 
- Equation 2: $$A^{[1]} = \sigma(Z^{[1]})$$
- Equation 3: $$Z^{[2]} = W^{[2]}A^{[1]} + b^{[2]}$$
- Equatoin 4: $$A^{[2]} = \sigma(Z^{[2]})$$

### 3. Geometric Nature of Result Matrices Z and A
Grasping the significance of the dimensions in matrices $Z$ and $A$ is a key factor in avoiding programming errors. Any output matrix $A^{[l]}$ or $Z^{[l]}$ adheres to the following coordinate positioning rule:
- **Horizontal axis (Columns):** Corresponds to scanning across different training data samples. Moving from the left column to the right column, one considers the first sample, the second sample, up to the $m$-th sample.
- **Vertical axis (Rows):** Corresponds to different neurons (hidden units) within the same layer. Scanning from top to bottom, one considers the value of the first neuron, the second neuron, etc.

**Illustrative example:** The element located at the top-left corner of matrix $A^{[1]}$ is exactly $a_1^{[1](1)}$. This is the activation value of the first hidden neuron computed on the first training sample.

## Explanation for Vectorized Implementation
### 1. Simplifying the Problem
To focus on the core nature of matrix multiplication, the initial assumption is to temporarily eliminate the bias term $b$ (i.e., treating $b = 0$).

Consequently, the linear computation equation at the first layer for a data sample $i$ is reduced to:
$$
z^{[1](i)} = W^{[1]} \cdot x^{(i)}
$$

### 2. Analyzing the Matrix Multiplication Mechanism
Assume the training dataset consists of 3 independent samples: $x^{(1)}, x^{(2)}$, and $x^{(3)}$. When computing each sample discretely, we obtain 3 corresponding output column vectors:
- $z^{[1](1)} = W^{[1]} \cdot x^{(1)}$
- $z^{[1](2)} = W^{[1]} \cdot x^{(2)}$
- $z^{[1](3)} = W^{[1]} \cdot x^{(3)}$

The vectorization technique requires stacking these 3 input vectors horizontally side-by-side to form the matrix $X$:
$$
X = \begin{bmatrix} | & | & | \\ x^{(1)} & x^{(2)} & x^{(3)} \\ | & | & | \end{bmatrix}
$$
According to the rules of matrix multiplication in linear algebra, when multiplying the weight matrix $W^{[1]}$ by the matrix $X$, the multiplication is automatically distributed to each column of $X$. The returned result is a new matrix, where each column is the product of $W^{[1]}$ and the corresponding column of $X$.

The mathematical operation unfolds as follows:
$$
W^{[1]} \cdot X = W^{[1]} \cdot \begin{bmatrix} | & | & | \\ x^{(1)} & x^{(2)} & x^{(3)} \\ | & | & | \end{bmatrix} = \begin{bmatrix} | & | & | \\ W^{[1]}x^{(1)} & W^{[1]}x^{(2)} & W^{[1]}x^{(3)} \\ | & | & | \end{bmatrix}
$$

Since $W^{[1]}x^{(i)}$ is exactly the definition of $z^{[1](i)}$, the resulting matrix perfectly matches the definition of the matrix $Z^{[1]}$ established in the previous lesson:
$$
Z^{[1]} = \begin{bmatrix} | & | & | \\ z^{[1](1)} & z^{[1](2)} & z^{[1](3)} \\ | & | & | \end{bmatrix}
$$
This proves that stacking the input vectors into columns preserves the order and generates the corresponding output vectors stacked in vertical columns. This principle applies consistently to all subsequent layers, such as calculating $A^{[1]}, Z^{[2]}$, and $A^{[2]}$.

### 3. The Role of Broadcasting
When restoring the bias term $b \neq 0$, the complete equation is:
$$
Z^{[1]} = W^{[1]}X + b^{[1]}
$$
Here, $W^{[1]}X$ is a matrix with dimensions $(n^{[1]}, m)$, whereas $b^{[1]}$ is merely a column vector with dimensions $(n^{[1]}, 1)$.
Python's Broadcasting mechanism automatically copies this vector $b^{[1]}$ horizontally $m$ times. As a result, the bias is added independently and accurately to each column of the matrix $Z^{[1]}$, completing the computational process without the need to manually construct a matrix for $b$.

<img src="../images/Week03/image6.png" width=600>

### 4. Symmetry of Neural Network Equations
To make the system of equations absolutely uniform, a convention is established: The input feature matrix $X$ is considered the activation matrix of layer 0, meaning $X = A^{[0]}$.

Thereby, the forward propagation equations for any layer $l$ in the neural network follow a single, unified mathematical format:
$$
Z^{[l]} = W^{[l]}A^{[l-1]} + b^{[l]}
$$
$$
A^{[l]} = g(Z^{[l]})
$$
(Where $g$ is the activation function)

The nature of Deep Neural Networks with dozens or hundreds of layers is essentially the repetition of this symmetric block of equations.

## Activation Functions
### I. Default Activation Functions in Deep Learning
Activation functions play a crucial role in introducing non-linearity to neural networks, enabling models to learn complex data representations. The choice of activation function directly affects the convergence speed and stability of the training process.

#### 1. Sigmoid Function
The Sigmoid function has an S-shaped curve graph, compressing all input values into the range between 0 and 1.
- **Mathematical nature:**
  $$
  a = \sigma(z) = \frac{1}{1 + e^{-z}}
  $$
- **Advantages:** The output is bounded between 0 and 1, which is ideal for representing probabilities. The curve is smooth and continuous.
- **Disadvantages:** It is prone to the vanishing gradient problem when the input value $z$ is extremely large or small. Additionally, the output is not zero-centered, which causes the Gradient Descent algorithm to optimize in a zig-zag path, slowing down the convergence rate.
- **Application:** Only used at the output layer of binary classification problems. It should absolutely not be used for hidden layers.

<img src="../images/Week03/image8.png" width=600>

#### 2. Tanh (Hyperbolic Tangent) Function
The Tanh function is a shifted version of the Sigmoid function, passing through the origin and having a value range from -1 to 1.
- **Mathematical nature:**
  $$
  a = \tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}
  $$
- **Advantages:** The output is zero-centered, helping the data to be naturally normalized when passed to the next layer. The convergence speed is significantly faster than Sigmoid.
- **Disadvantages:** It is still limited by the vanishing gradient phenomenon at both ends of the graph.
- **Application:** It used to be the standard choice for hidden layers before ReLU was introduced. It is still used in Recurrent Neural Networks (RNNs).

<img src="../images/Week03/image9.png" width=600>

#### 3. ReLU (Rectified Linear Unit) Function
ReLU is the standard and most popular activation function in modern Deep Learning architectures.
- **Mathematical nature:**
  $$
  a = \max(0, z)
  $$
- **Advantages:** It completely overcomes the vanishing gradient problem in the positive domain because the derivative is always 1 when $z > 0$. The computational cost is extremely low as it only uses a simple comparison operation.
- **Disadvantages:** It suffers from the "dying ReLU" problem. When $z < 0$, the derivative immediately becomes 0. The weights of neurons that fall into this state will never be updated again.
- **Application:** The default choice and top priority for all hidden layers.

<img src="../images/Week03/image10.png" width=600>

#### 4. Leaky ReLU Function
Leaky ReLU is a variant designed to overcome the dying neuron drawback of the original ReLU.
- **Mathematical nature:**
  $$
  a = \max(\alpha z, z)
  $$
  where $\alpha$ is usually a small value like $0.01$.
- **Advantages:** It maintains a small gradient in the negative domain, giving neurons a chance to continue learning and updating weights even when receiving input values less than 0.
- **Disadvantages:** It requires the programmer to manually tune an additional hyperparameter $\alpha$.
- **Application:** Used as a replacement for ReLU in hidden layers when the model experiences stagnation due to too many deactivated neurons.

<img src="../images/Week03/image11.png" width=600>

#### 5. Softmax Function
Softmax does not operate on a single neuron but normalizes the entire output vector of a layer into a probability distribution.
- **Mathematical nature:** For a layer with $K$ neurons, the activation value of the $i$-th neuron is:
  $$
  a_i = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
  $$
- **Advantages:** Amplifies the differences between elements. It ensures that all output values fall within the range of 0 to 1 and their sum always exactly equals 1.
- **Disadvantages:** High computational cost due to the use of complex exponential functions across the entire vector.
- **Application:** The mandatory choice for the output layer of multi-class classification problems.

### II. The Problem of Vanishing and Exploding Gradients
When building deep neural networks (with many hidden layers), backpropagating errors to update weights frequently encounters two severe mathematical barriers.

#### 1. Mathematical Nature (Chain Rule)
To calculate the derivative of the loss function $\mathcal{L}$ with respect to the first layer's weights $W^{[1]}$, the backpropagation algorithm performs consecutive multiplications of local derivatives from the last layer back to the first layer:
$$
\frac{\partial \mathcal{L}}{\partial W^{[1]}} = \frac{\partial \mathcal{L}}{\partial a^{[L]}} \cdot \frac{\partial a^{[L]}}{\partial z^{[L]}} \cdot \frac{\partial z^{[L]}}{\partial a^{[L-1]}} \dots \frac{\partial z^{[2]}}{\partial a^{[1]}} \cdot \frac{\partial a^{[1]}}{\partial W^{[1]}}
$$

#### 2. Vanishing Gradient Phenomenon
- **Definition:** Occurs when the local derivative values are less than 1. Multiplying a series of these fractions together causes the product to decay exponentially and rapidly approach 0.
- **Impact:** The gradients at the initial layers are nearly 0. The network stagnates; the foundational layers cannot update their weights and fail to learn basic features. The deep model then performs no better than a shallow model.
- **Causes:** Often due to the overuse of the Sigmoid activation function or initializing weights too small.

#### 3. Exploding Gradient Phenomenon
- **Definition:** Occurs when the local derivative values are greater than 1. Consecutive multiplications cause the product to grow exponentially and approach infinity.
- **Impact:** The Gradient Descent algorithm takes steps that are too large, overshooting the local optimum. The training process becomes unstable, the loss function fluctuates wildly, or encounters numerical overflow errors (NaN), leading to model collapse.
- **Causes:** Primarily due to initializing the weight matrix $W$ with random values that are too large.

### III. System of Solutions
To successfully build highly stable deep neural networks, the field of Deep Learning applies a combination of the following methods:
- **Optimizing Activation Functions:** Completely replace Sigmoid and Tanh with ReLU or Leaky ReLU in hidden layers. The derivative of 1 for ReLU in the positive domain plays a core role in maintaining the strength of the gradient signal as it backpropagates through hundreds of layers.
- **Precise Weight Initialization:** Do not use standard random distributions. Instead, apply He Initialization when using the ReLU function and Xavier (Glorot) Initialization when using the Tanh function. These methods strictly control the variance of the weights based on the number of input neurons, ensuring the signal is not abnormally amplified or diminished.
- **Batch Normalization:** Directly normalizes the linear value $Z$ at each hidden layer so that it has a mean of 0 and a variance of 1 before passing it through the activation function. This method forces the data into the region with the best gradients, thoroughly preventing both exploding and vanishing phenomena.
- **Gradient Clipping:** A direct solution for the exploding gradient phenomenon. When the gradient value exceeds an established safe threshold, the system proactively scales it back down to that threshold, ensuring the weights' structure is not destroyed by excessively large update steps.
- **Residual Networks (ResNet) Architecture:** Redesign the network structure by adding Skip Connections. Data is directly added from previous layers to deeper layers, creating a straight path for the gradient to flow back to the initial layers without interference from consecutive matrix multiplications.

## Why do you need Non-Linear Activation Function?
### 1. The Collapse of Linear Hidden Layers
Suppose we remove the non-linear functions and replace them with a linear activation function (also known as the identity function), where the output equals the input: $g(z) = z$. In this case, the activation value $a$ will be exactly equal to the linear value $z$.

Let's consider the forward propagation process on a 2-layer neural network:
- **Hidden layer 1 equation:** $a^{[1]} = z^{[1]} = W^{[1]}x + b^{[1]}$
- **Output layer 2 equation:** $a^{[2]} = z^{[2]} = W^{[2]}a^{[1]} + b^{[2]}$

Substituting the equation of layer 1 into the equation of layer 2, we have:
$$
a^{[2]} = W^{[2]}(W^{[1]}x + b^{[1]}) + b^{[2]}
$$

Expanding this expression, we get:
$$
a^{[2]} = (W^{[2]}W^{[1]})x + (W^{[2]}b^{[1]} + b^{[2]})
$$

According to the fundamental rules of linear algebra:
- The product of the two matrices $W^{[2]}W^{[1]}$ will generate a new weight matrix, which we can call $W'$.
- The combination of the operations $W^{[2]}b^{[1]} + b^{[2]}$ will generate a new bias vector, which we can call $b'$.

The final generalized expression is reduced to:
$$
a^{[2]} = W'x + b'
$$

**Core Conclusion:** The reduced equation $a^{[2]} = W'x + b'$ is exactly the standard equation of a basic Linear Regression model.

This proves that the composition of multiple linear functions inherently produces only a single linear function. Without a non-linear activation function, stacking dozens or hundreds of hidden layers is completely meaningless, as the computational capacity of that entire deep network is no superior to a single-layer linear model.

### 2. The Significance of Non-linearity
For a neural network to solve complex real-world problems (such as image recognition, natural language processing, or datasets that cannot be separated by a straight line), the model must have the ability to bend and reshape the coordinate space.

Functions like Sigmoid, Tanh, or ReLU are the tools that create this non-linear bending at each neuron, enabling the hidden layers to extract complex machine learning features.

### 3. When is it permissible to use a linear function?
Although hidden layers must strictly use non-linear functions, the linear activation function $g(z) = z$ still has one specific application: **The output layer of a Regression problem**.
- **Regression Problems:** When you need to predict a continuous value, such as housing prices. The target variable $y$ is a real number $\mathbb{R}$.
- **Output Layer Design:** If the Sigmoid function is used in the final layer, the predicted house price $\hat{y}$ will be erroneously constrained between $0$ and $1$. Therefore, the output layer needs to use a linear function so the model can return a real number freely ranging from $-\infty$ to $+\infty$.
- **Practical Note:** If the predicted data has a non-negative characteristic (like house prices always being $\ge 0$), you could even use the ReLU function at the output layer to ensure the returned result is always logically valid.

## Derivatives of Activation Functions
### 1. Derivative of the Sigmoid Function
Suppose the activation value at a neuron is $a = g(z) = \sigma(z)$. Using basic calculus rules, the derivative of the Sigmoid function is reduced to an incredibly elegant expression, depending only on the output value $a$ itself:
$$
g'(z) = a(1 - a)
$$
<img src="../images/Week03/image12.png" width=400>

**Graph behavior analysis:**
- When $z$ has a very large value (e.g., $z = 10$), the value $a$ approaches 1. The derivative is then approximately $g'(z) \approx 1(1 - 1) = 0$.
- When $z$ has a very small value (e.g., $z = -10$), the value $a$ approaches 0. The derivative is then approximately $g'(z) \approx 0(1 - 0) = 0$.
- When $z = 0$, the value $a$ is exactly in the middle at 0.5. The derivative reaches its maximum at $g'(z) = 0.5(1 - 0.5) = 0.25$.

**Practical advantage:** This expression is highly optimized for computer resources. During forward propagation, the neural network has already calculated the value $a$ and stored it in memory. Therefore, the system can immediately reuse the variable $a$ to calculate the derivative during backpropagation without having to recalculate the expensive exponential function.

### 2. Derivative of the Tanh (Hyperbolic Tangent) Function
Similar to Sigmoid, for the Tanh function with an activation value $a = g(z) = \tanh(z)$, its derivative is also calculated directly through the square of the output value:
$$
g'(z) = 1 - a^2
$$
<img src="../images/Week03/image13.png" width=400>

**Graph behavior analysis:**
- When $z$ is very large or very small, the value $a$ will approach the maximum of 1 or the minimum of -1. When squared, $a^2$ always approaches 1, leading to a derivative $g'(z) \approx 1 - 1 = 0$. This phenomenon is the cause of the vanishing gradient at the two tails of the graph.
- When $z = 0$, the value $a = 0$. The derivative at this point reaches its maximum, which is $g'(z) = 1 - 0 = 1$. The slope of 1 right at the origin serves as a fulcrum that helps the Tanh function backpropagate errors much more effectively than the 0.25 maximum of the Sigmoid function.

### 3. Derivative of the ReLU and Leaky ReLU Functions
The ReLU (Rectified Linear Unit) function is defined as $g(z) = \max(0, z)$. Since the graph is composed of two straight lines, the derivative of ReLU is a piecewise function:
- If $z < 0$: $g'(z) = 0$
- If $z > 0$: $g'(z) = 1$

For the Leaky ReLU variant defined as $g(z) = \max(0.01z, z)$, the derivative operates on a similar principle but retains a very small slope for the negative domain to prevent the dying neuron problem:
- If $z < 0$: $g'(z) = 0.01$
- If $z > 0$: $g'(z) = 1$

**Exception handling at the breaking point:**
In pure calculus, the derivative exactly at the point $z = 0$ is undefined. However, in practical software programming, the probability of a floating-point real variable having an absolute exact value of 0.000000 is extremely rare.

To completely resolve this mathematical loophole, experts use the concept of a sub-gradient, allowing the system to automatically assign a value of $g'(0) = 0$ or $g'(0) = 1$ in the source code. This setup ensures that the Gradient Descent algorithm always converges stably.

### 4. Derivative of the Softmax Function
Completely different from the functions above, which only operate on individual independent neurons, the Softmax function calculates the probability of a neuron based on the aggregate of all neurons in the same layer.

With a layer having $K$ neurons, the Softmax equation at the $i$-th neuron is:
$$
a_i = \frac{e^{z_i}}{\sum_{k=1}^{K} e^{z_k}}
$$

Due to this interconnected property, the derivative of Softmax is a Jacobian matrix of partial derivatives. We must consider two distinct mathematical cases:
- **Case 1 (Derivative with respect to itself, $i = j$):** Using the quotient rule for derivatives, the expression simplifies to $\frac{\partial a_i}{\partial z_i} = a_i(1 - a_i)$. This expression shares a perfect mathematical similarity with the derivative of the Sigmoid function.
- **Case 2 (Derivative with respect to another element, $i \neq j$):** When $z_j$ changes, the denominator sum changes, causing $a_i$ to change accordingly. The expression carries a negative sign, reflecting mutually exclusive competition: $\frac{\partial a_i}{\partial z_j} = -a_i a_j$.

**Generalization using the Kronecker Delta:**

To allow computers to calculate this without complex branching commands, the two cases above are combined into a single equation using the Kronecker Delta function $\delta_{ij}$ (equal to 1 when $i=j$ and equal to 0 when $i \neq j$):
$$
\frac{\partial a_i}{\partial z_j} = a_i(\delta_{ij} - a_j)
$$

**Optimization at the output layer:**

In practical model development, the Softmax function is always tightly coupled with the Cross-Entropy loss function. According to the chain rule, when multiplying the derivative of Cross-Entropy by the Jacobian matrix of Softmax, a special mathematical cancellation occurs. All fractions, exponentials, and summation signs automatically cancel each other out, leaving behind a highly streamlined core backpropagation equation that processes at lightning speed:
$$
dZ^{[L]} = A^{[L]} - Y
$$
Where $A^{[L]}$ is the Softmax prediction vector and $Y$ is the actual label vector.

## Gradient Descent for Neural Networks
### 1. Initializing Matrix Dimensions
Consider a neural network solving a Binary Classification problem with the following basic architecture:
- Number of input features: $n^{[0]} = n_x$
- Number of hidden layer neurons: $n^{[1]}$
- Number of output layer neurons: $n^{[2]} = 1$

For matrix multiplications to be valid, the weight matrices $W$ and bias vectors $b$ must be initialized with precise dimensions:
- $W^{[1]}$: dimension $(n^{[1]}, n^{[0]})$
- $b^{[1]}$: dimension $(n^{[1]}, 1)$
- $W^{[2]}$: dimension $(n^{[2]}, n^{[1]})$
- $b^{[2]}$: dimension $(n^{[2]}, 1)$

**Important Note:** The neural network must initialize the weight matrices $W$ with random values instead of zeros to break symmetry. The bias vectors $b$ can safely be initialized to zero.

### 2. Forward Propagation
This process receives the input data matrix $X$ and computes sequentially through the layers to output the final prediction $\hat{Y} = A^{[2]}$. All operations are vectorized to process all $m$ data samples simultaneously.
- **At Hidden Layer 1:** Typically uses the ReLU or Tanh activation function.
  $$
  Z^{[1]} = W^{[1]}X + b^{[1]}
  $$
  $$
  A^{[1]} = g^{[1]}(Z^{[1]})
  $$
- **At Output Layer 2:** Strictly uses the Sigmoid function for binary classification problems.
  $$
  Z^{[2]} = W^{[2]}A^{[1]} + b^{[2]}
  $$
  $$
  A^{[2]} = g^{[2]}(Z^{[2]})
  $$
After calculating $A^{[2]}$, the system will compare it with the actual labels $Y$ through the Loss Function $\mathcal{L}$ to evaluate the model's error.

### 3. Backpropagation
This is the heart of the Gradient Descent algorithm. The system calculates the error propagating backward from the output layer to the hidden layer to determine the necessary adjustments for the parameters. The entire process is encapsulated in the following 6 mathematical equations:

**At Output Layer 2:**
$$
dZ^{[2]} = A^{[2]} - Y
$$
$$
dW^{[2]} = \frac{1}{m} dZ^{[2]} (A^{[1]})^T
$$
$$
db^{[2]} = \frac{1}{m} \text{np.sum}(dZ^{[2]}, \text{axis}=1, \text{keepdims=True})
$$

**At Hidden Layer 1:**
$$
dZ^{[1]} = (W^{[2]})^T dZ^{[2]} * g^{[1]'}(Z^{[1]})
$$
$$
dW^{[1]} = \frac{1}{m} dZ^{[1]} X^T
$$
$$
db^{[1]} = \frac{1}{m} \text{np.sum}(dZ^{[1]}, \text{axis}=1, \text{keepdims=True})
$$

(The $*$ symbol denotes element-wise multiplication)

### 4. Mathematical Derivation of the Core Equation using the Chain Rule
Equation number 4 for calculating $dZ^{[1]}$ looks quite complex but is actually an inevitable consequence of the Chain Rule in calculus. Our goal is to find the rate of change of the loss function $\mathcal{L}$ with respect to the variable $Z^{[1]}$:
$$
dZ^{[1]} = \frac{\partial \mathcal{L}}{\partial Z^{[1]}}
$$
The variable $Z^{[1]}$ goes through a forward propagation chain: $Z^{[1]} \rightarrow A^{[1]} \rightarrow Z^{[2]} \rightarrow \mathcal{L}$. Therefore, we must compute backward through each intermediate station:

**Station 1:** From $Z^{[2]}$ back to $A^{[1]}$ (Calculating $dA^{[1]}$)

Based on the forward propagation equation $Z^{[2]} = W^{[2]}A^{[1]} + b^{[2]}$, we calculate the derivative of $\mathcal{L}$ with respect to $A^{[1]}$:
$$
dA^{[1]} = \frac{\partial \mathcal{L}}{\partial A^{[1]}} = \frac{\partial \mathcal{L}}{\partial Z^{[2]}} \cdot \frac{\partial Z^{[2]}}{\partial A^{[1]}}
$$
- The component $\frac{\partial \mathcal{L}}{\partial Z^{[2]}}$ is exactly the error term $dZ^{[2]}$.
- The component $\frac{\partial Z^{[2]}}{\partial A^{[1]}}$ returns the weight matrix $W^{[2]}$. To preserve the dimensional validity of the matrices, we must use the transposed matrix $(W^{[2]})^T$.
- Station 1 Result: $dA^{[1]} = (W^{[2]})^T dZ^{[2]}$

**Station 2: From $A^{[1]}$ back to $Z^{[1]}$ (Calculating $dZ^{[1]}$)**

Based on the activation equation $A^{[1]} = g^{[1]}(Z^{[1]})$, we complete the journey back to $Z^{[1]}$:
$$
dZ^{[1]} = \frac{\partial \mathcal{L}}{\partial Z^{[1]}} = \frac{\partial \mathcal{L}}{\partial A^{[1]}} \cdot \frac{\partial A^{[1]}}{\partial Z^{[1]}}
$$
- The component $\frac{\partial \mathcal{L}}{\partial A^{[1]}}$ is exactly the $dA^{[1]}$ we just found at Station 1.
- The component $\frac{\partial A^{[1]}}{\partial Z^{[1]}}$ is the derivative of the activation function $g^{[1]'}(Z^{[1]})$. Since the activation function is applied independently to each element, we use the element-wise multiplication operator $*$.
- Assembling them, we get the final answer:
  $$
  dZ^{[1]} = (W^{[2]})^T dZ^{[2]} * g^{[1]'}(Z^{[1]})
  $$

### 5. Technical Notes on Python (NumPy) Programming
In the 6 backpropagation equations, calculating $db^{[1]}$ and $db^{[2]}$ requires using the np.sum function. Professor Andrew Ng emphasizes the mandatory use of the following two parameters to avoid data structure errors:
- `axis=1`: Specifies summing horizontally, aggregating the entire error across all $m$ training examples.
- `keepdims=True`: This is a shield protecting the matrix structure. If this parameter is missing, Python will automatically compress the result into a Rank 1 Array of the form `(n,)`, causing severe bugs in subsequent steps. This parameter forces Python to maintain the standard column vector structure `(n, 1)`.

### 6. Gradient Descent Update
Once all the derivative matrices $dW$ and $db$ are obtained, the model will proceed to adjust the parameters based on the learning rate $\alpha$:
$$
W^{[1]} = W^{[1]} - \alpha \cdot dW^{[1]}
$$
$$
b^{[1]} = b^{[1]} - \alpha \cdot db^{[1]}
$$
$$
W^{[2]} = W^{[2]} - \alpha \cdot dW^{[2]}
$$
$$
b^{[2]} = b^{[2]} - \alpha \cdot db^{[2]}
$$

The entire cycle from step 2 to step 6 will be repeated by the system thousands of times until the loss function converges to its minimum level, signaling the completion of the training process.

## Random Initialization
### I. The Core Problem: Collapse due to Symmetry
For simple models like Logistic Regression, you can initialize all weights to 0 initially, and the model will still converge. However, if this method is applied to a neural network with a hidden layer, the Gradient Descent algorithm will fail completely.

Suppose you have a neural network with inputs $n^{[0]} = 2$ and a hidden layer $n^{[1]} = 2$. The weight matrix $W^{[1]}$ will have a dimension of $(2, 2)$. If you initialize $W^{[1]}$ and the bias vector $b^{[1]}$ to 0:
$$
W^{[1]} = \begin{bmatrix} 0 & 0 \\ 0 & 0 \end{bmatrix}, b^{[1]} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

At this point, the neural network encounters a severe architectural flaw:
- **Forward Propagation:** Both neurons in the hidden layer receive the exact same input data and multiply it by a weight matrix of 0. As a result, their activation values are always equal ($a_1^{[1]} = a_2^{[1]}$) for every data sample.
- **Backpropagation:** Due to symmetry, the error propagated back to these two neurons is also completely identical ($dz_1^{[1]} = dz_2^{[1]}$). Consequently, the derivative matrix for updates, $dW^{[1]}$, will have rows with the exact same values.

When Gradient Descent updates the weights ($W^{[1]} = W^{[1]} - \alpha \cdot dW^{[1]}$), the rows of matrix $W$ will still change by equal amounts. No matter how many thousands of iterations you train the model, the two hidden neurons will always remain replicas of each other and perform the exact same calculation. The consequence is that a vast network degenerates, and its representational power is equivalent to only a single neuron.

### II. Common Weight Initialization Techniques
To break the symmetry and optimize the data flow, computer science has developed the following initialization methods:

#### 1. Zero Initialization
- **Nature:** Assigns a value of $0$ to all parameters.
- **Application Strategy:** Absolutely do not use this for the weight matrix $W$ in hidden layers to avoid the symmetry error analyzed above. This technique is only safe and valid when used to initialize the bias vector $b$.

#### 2. Random Initialization
- **Nature:** Assigns random values drawn from a Normal distribution or a Uniform distribution, then multiplies them by a very small constant (e.g., 0.01) to force the weights to remain close to 0.
- **Application Strategy:** Should only be used for very shallow neural networks (with 1 hidden layer) or as a baseline model for quick testing. For deep networks, this method is highly risky as it does not control the variance, easily leading to instability during backpropagation.

#### 3. Xavier / Glorot Initialization
This is a major breakthrough in Deep Learning, automatically balancing the variance of the weights based on the number of input and output neurons. The signal will not be abnormally amplified or diminished as it passes through multiple hidden layers.
- **Xavier Normal:** Weights are randomly drawn from a Normal distribution $\mathcal{N}(0, \sigma)$ with variance:
  $$
  \sigma = \sqrt{\frac{2}{fan\_in + fan\_out}}
  $$
- **Xavier Uniform:** Weights are randomly drawn from a Uniform distribution:
  $$
  W \sim U\left[-\sqrt{\frac{6}{fan\_in + fan\_out}}, \sqrt{\frac{6}{fan\_in + fan\_out}}\right]
  $$
- **Application Strategy:** The preferred and optimal choice when your hidden layers use the Sigmoid or Tanh activation functions.

#### 4. He Initialization
The ReLU function has the characteristic of completely eliminating negative values, neutralizing half of the data space. Therefore, the original Xavier method can no longer maintain optimal variance. He Initialization was created to overcome this by doubling the variance and focusing only on the number of inputs.
- **He Normal:** Weights are randomly drawn from a Normal distribution $\mathcal{N}(0, \sigma)$ with:
  $$
  \sigma = \sqrt{\frac{2}{fan\_in}}
  $$
- **He Uniform:** Weights are randomly drawn from a Uniform distribution:
  $$
  W \sim U\left[-\sqrt{\frac{6}{fan\_in}}, \sqrt{\frac{6}{fan\_in}}\right]
  $$
- **Application Strategy:** The mandatory technique to achieve the highest performance when the network architecture uses the ReLU activation function or its variants (Leaky ReLU, PReLU).
