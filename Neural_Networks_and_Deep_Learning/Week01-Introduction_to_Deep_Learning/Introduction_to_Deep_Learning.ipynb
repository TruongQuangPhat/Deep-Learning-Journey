## What is a Neural Networks?

### 1. Single Neuron and the ReLU Activation Function
Basic linear regression models often face limitations in real-world problems with natural boundaries. For instance, house prices cannot be negative. If only a linear line is used for prediction, the model might return illogical values.

To solve this problem, neural networks use a mathematical transformation to restrict the output. Instead of a straight line running through the horizontal axis, the prediction line is bent so that the minimum value always stops at 0. The mathematical function performing this operation is called the **Rectified Linear Unit (ReLU)**. Mathematically, the ReLU function takes the maximum value between 0 and the input: $f(x) = \max(0, x)$.

A single neuron in the network is a computational unit that takes an input $x$ (house size), applies the ReLU function, and outputs a result $y$ (predicted house price).

### 2. Multi-feature Neural Network Architecture
A complete neural network is built by stacking and connecting numerous single neurons together. The true power of a neural network emerges when we provide it with a multidimensional space containing multiple input features.

In the house pricing problem, instead of using only the size, the input set $x$ can include four features:
- Size
- Number of bedrooms
- Zip code
- Average income level of the area

The combination of these raw features forms more abstract concepts that are decisive to the value of the house. For example, the size and the number of bedrooms determine the capacity for the family size; the zip code indicates walkability; the zip code combined with the income level reflects the quality of nearby schools.

### 3. Hidden Layers and Dense Connections
A neural network structure consists of an **Input Layer**, **Hidden Layers**, and an **Output Layer**.

A special characteristic of a densely connected (Fully Connected) architecture is that every neuron in the previous layer transmits signals to every neuron in the subsequent layer. Humans do not need to (and should not) intervene to specify which hidden neuron represents family size and which represents school quality. By receiving all input features, the neural network will automatically calculate, learn, and decide how to represent these latent factors in the middle layers to serve the ultimate goal of accurately predicting the house price $y$.

### 4. Supervised Learning Mechanism
Neural networks are most effective in supervised learning problems. This process requires a training dataset containing labeled data pairs, including the input $x$ and the actual result $y$.

When provided with a sufficient amount of training data, through continuous calculation and weight adjustment steps, the neural network has the ability to automatically discover the complex mapping function from $x$ to $y$. It converts raw data into highly accurate predictions without requiring programmers to manually write conditional rule sets.

## Supervised Learning with Neural Networks
### 1. Supervised Learning and Economic Motivation
Despite numerous hyped hypotheses about the capabilities of artificial intelligence, so far, almost all of the practical economic value brought by neural networks stems from one core method: **Supervised Learning**.

Essentially, this process involves finding an optimal mathematical function to map from an input dataset $x$ to an output result $y$. Skillfully determining what $x$ and $y$ should be dictates the success or failure of an application.

Some of the most profitable and breakthrough applications include:
- **Online Advertising:** The input $x$ is user information and ad content, and the output $y$ is the probability of a user clicking. The neural network's accurate prediction capability brings enormous profits to tech companies.
- **Computer Vision:** The input $x$ is an image, and the output $y$ is a classification label (e.g., an index from 1 to 1000 to identify objects in the image).
- **Speech Recognition & Machine Translation:** The input $x$ is an audio clip or an English sentence, and the output $y$ is the transcribed text or the corresponding Chinese sentence.
- **Autonomous Driving:** The input $x$ is an image from a camera and radar data, and the output $y$ is the exact position of other vehicles on the road.

### 2. Specialized Neural Network Architectures
No single neural network can solve every problem. Depending on the characteristics of the input data, engineers will apply different architectures:
- **Standard Neural Networks (Standard NN):** Often composed of densely connected layers (Fully Connected Layers). This architecture proves highly effective for real estate pricing problems or online advertising recommendation systems.
- **Convolutional Neural Networks (CNN):** This is the gold standard for image data. By using convolutional filters, CNNs preserve the spatial structure of pixels to recognize everything from basic features (edges, corners) to complex objects.
- **Recurrent Neural Networks (RNN):** Specifically designed for Sequence Data that has a temporal or sequential nature. Typical examples are audio signals (1D time series) or natural language (where the order of words determines semantics).
- **Hybrid Architecture:** In complex systems like autonomous vehicles, engineers often combine multiple architectures, such as using a CNN to process camera images and connecting it with other components to process radar data.

### 3. Structured vs. Unstructured Data
**Structured Data:**
- This is data clearly organized in the form of tables or databases.
- Each feature has a specific meaning, for example: house size, number of bedrooms, or user age.
- In reality, the vast majority of short-term economic value and corporate profits today are still generated from analyzing these massive structured databases.

**Unstructured Data:**
- These are raw forms of information such as audio, pixels, or words in free text.
- Previously, it was very difficult for computers to extract meaning from this type of data. However, thanks to Deep Learning, this barrier has been broken. The fact that artificial neural networks can recognize an image of a cat or understand human speech are advancements that attract significant media attention, opening up countless new application opportunities.

## Why Deep Learning taking off?
### 1. The Explosion of the Big Data Era
The first and most decisive factor is the scale of data. In traditional Machine Learning algorithms (such as Support Vector Machines or Logistic Regression), predictive performance tends to improve as more data is provided, but it quickly reaches a state of saturation. At that point, the performance curve flattens out; the algorithm is unable to extract more value regardless of how much massive data you feed into it.

In contrast, Deep Neural Network architectures have an unlimited capacity to absorb and learn from data.
- If you train a small neural network, the performance will be slightly better.
- Training a medium-sized neural network, the performance continues to increase.
- When you train an extremely large neural network (many hidden layers, many parameters) on a massive dataset, its performance will consistently surpass older methods.

The rise of the Internet, smartphones, mobile applications, and IoT (Internet of Things) sensors has pushed human society into a digitalized era. The massive amount of digital data generated every day (audio, images, text) is the abundant fuel that allows Deep Learning models to maximize their power.

### 2. Tremendous Advancements in Computational Power
Training a massive neural network with millions or even billions of parameters requires an enormous amount of computational power. The rapid development of specialized hardware, particularly Graphics Processing Units (GPUs) and high-speed network connections, has overcome this barrier.

Powerful computational capacity not only helps models process larger amounts of data but also completely transforms the iterative workflow of AI researchers and engineers:
- The process of building a Deep Learning model is highly iterative: **Idea -> Code -> Experiment -> Analyze results -> Adjust parameters**.
- Previously, training a model could take months, causing the idea-testing process to stall. Today, with modern computational power, this time can be reduced to a few days, a few hours, or even 10 minutes. This allows engineers to execute more iteration cycles, explore more complex architectures, and optimize models much faster.

### 3. Algorithmic Innovations
Although the mathematical foundations have existed for a long time, clever algorithmic improvements over the past decade have helped solve severe bottlenecks, particularly in optimizing the learning speed of neural networks.

A classic example given by Andrew Ng is the shift from the **Sigmoid** activation function to the **ReLU (Rectified Linear Unit)** function:
- **Sigmoid function:** Has a critical flaw at the two extremes of the graph, where the slope (gradient) is almost zero. When applying the Gradient Descent algorithm, a gradient of zero prevents parameters from updating, leading to the **Vanishing Gradient** problem. This makes the network's learning process extremely slow or even stops it entirely.
- **ReLU function:** Unlike Sigmoid, for all positive input values, the derivative (slope) of the ReLU function is always a constant equal to 1. This helps eliminate the vanishing gradient problem, allowing the Gradient Descent algorithm to run smoothly and converge many times faster.
