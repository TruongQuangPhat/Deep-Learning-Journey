## Binary Classification
### 1. Vectorization Technique

When working with large datasets, using repetitive loops to process each data sample significantly reduces computational performance. In Deep Learning programming, the vectorization technique is mandatory. This technique allows processing the entire training set simultaneously through matrix operations, helping to optimize the learning speed of the model.

Furthermore, the neural network training process is always organized into two distinct mathematical steps:
- **Forward Propagation:** Calculating the predicted value from the input data.
- **Backward Propagation:** Calculating the derivatives to update the parameters.

To get accustomed to this mindset, the Logistic Regression problem is used as a simple neural network model consisting of a single neuron for illustration.

### 2. Binary Classification Setup

The goal of this problem is to predict an output that has only two possibilities. A typical example is image recognition:
- **Input:** An image.
- **Output $y$:** A binary label, where 1 means the image contains a cat and 0 means the image does not contain a cat.

### 3. Image Data Representation
Computers receive images in the form of pixel matrices. An RGB color image consists of three matrices corresponding to the three color channels: Red, Green, and Blue.

If an image has a resolution of 64 x 64 pixels, it will be represented by three 64 x 64 matrices.
To feed this image into a training model, all values of these three matrices will be flattened into a single column vector. This is called the input feature vector and is denoted as $x$.

The size of vector $x$, denoted as $n_x$ or $n$, will be the total number of pixels multiplied by the three color channels:
$$
n_x = 64 \times 64 \times 3 = 12,288
$$
Thus, each image is now represented by a column vector with a length of 12,288 dimensions.

### 4. Notation Standards
To ensure the algorithm implementation via code is as smooth and accurate as possible, standardizing mathematical notation is particularly important:
- **A training example:** Represented by the pair $(x, y)$, where $x$ is the feature vector and $y$ is the target label belonging to the set of 0 or 1.
- **Number of examples:** $m$ represents the total number of data samples in the training set or test set.
- **Sample index:** Data samples are marked with an index placed in parentheses as a superscript. For example, $(x^{(1)}, y^{(1)})$ is the first sample, and $(x^{(m)}, y^{(m)})$ is the $m$-th sample.

### 5. Building the Input Matrix (X) and Output Matrix (Y)
To eliminate loops, all $m$ data samples will be stacked into large matrices.

**Input Feature Matrix $X$:**

Instead of stacking the $x$ vectors horizontally, in Deep Learning, the vectors from 1 to $m$ will be stacked as the columns of matrix $X$.
- Number of rows in $X$: $n_x$
- Number of columns in $X$: $m$
- Format in Python: X.shape = (n_x, m)

**Output Label Matrix $Y$:**

Similarly, the labels from 1 to $m$ are also stacked into a row vector, equivalent to a matrix with 1 row and $m$ columns.
- Format in Python: Y.shape = (1, m)

## Logistic Regression
### 1. The Goal of Logistic Regression in Binary Classification
Logistic Regression is the core algorithm for binary classification problems, where the actual label $y$ only takes one of two values, 0 or 1.

When providing the model with an input feature vector $x$, the algorithm's goal is to compute a predicted value, denoted as $\hat{y}$. Mathematically, $\hat{y}$ is the probability that $y = 1$ given the input $x$. For example, if the problem is image recognition, $\hat{y}$ will tell you the exact probability (from 0 to 1) that the image contains a cat.

### 2. Limitations of Linear Functions and the Sigmoid Solution
If we calculate the output based on a simple linear combination like the Linear Regression algorithm, the formula would be:
$$
z = w^T x + b
$$
This approach completely fails in binary classification. The reason is that this linear combination can return deeply negative or extremely large positive values. A predicted value that is negative or greater than 1 is meaningless when representing a probability.

To overcome this, Logistic Regression requires passing this value $z$ through a non-linear transformation called the **Sigmoid** function.

<img src="../images/Week02/image3.png" width=400>

The formula for the Sigmoid function is defined as follows:
$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

The mathematical characteristics of the Sigmoid function make it a perfect filter for probabilities:
- When $z$ is an extremely large positive number, $e^{-z}$ will gradually approach 0. At this point, the denominator is approximately 1, causing the overall function $\sigma(z)$ to approach 1.
- When $z$ is an extremely large negative number, $e^{-z}$ becomes a massive number. The extremely large denominator will force the entire fraction $\sigma(z)$ to approach 0.
- The graph of the function intersects the vertical axis at exactly 0.5.

Thanks to this S-shaped curve, no matter how extreme the linear combination $z$ gets, the output $\hat{y}$ is always smoothly bounded between 0 and 1.

<img src="../images/Week02/image4.png" width=800>

### 3. Parameter Notation Conventions (w and b)
In the approach of this course, the parameter set of the neural network is divided into two independent components:
- The weight vector $w$: Has the same dimensionality $n_x$ as the input vector $x$.
- The bias $b$: Is a single real number.

Andrew Ng also clarifies the difference compared to some other academic conventions. In some machine learning literature, people often combine $w$ and $b$ into a single parameter vector $\theta$. To do that, they have to insert a dummy feature $x_0 = 1$ into the input vector, turning $x$ into a vector in the $\mathbb{R}^{n_x + 1}$ space.

However, when programming deep neural networks in practice, this grouping makes matrix operations cumbersome. Keeping the weight vector $w$ and the bias $b$ strictly independent will make the code structure in Python much clearer and easier to implement.

## Logistic Regression Cost Function
### 1. Goal and Notation Conventions
During the training process, the ultimate goal is to make the predicted value $\hat{y}$ as close to the actual label $y$ as possible.

To accurately represent this mathematically for each data sample in a training set of $m$ examples, we use a superscript index enclosed in parentheses $(i)$.

Specifically, for the $i$-th data example:
- The input is $x^{(i)}$
- The actual label is $y^{(i)}$
- The intermediate linear value is $z^{(i)} = w^T x^{(i)} + b$
- The predicted value is $\hat{y}^{(i)} = \sigma(z^{(i)})$

<img src="../images/Week02/image5.png" width=800>

### 2. Loss Function for a Single Training Example
The **Loss Function**, denoted as $L(\hat{y}, y)$, is tasked with measuring the discrepancy between the predicted value $\hat{y}$ and the true label $y$ on a **single training example**.

**Why not use the Squared Error function?**

In linear regression, it is common to use the squared error function $\frac{1}{2}(\hat{y} - y)^2$. However, in Logistic Regression, due to the introduction of the non-linear Sigmoid function, using the squared error would turn the optimization problem into a non-convex function.

A non-convex function has many local optima. If we apply the Gradient Descent algorithm to a non-convex function, the algorithm can easily get stuck in these false valleys and never find the global optimum.

**The Solution: The Logistic Loss Function (Log-Loss)**

To ensure the optimization problem is a bowl-shaped convex function, Logistic Regression utilizes a specific loss function:
$$
L(\hat{y}, y) = -(y \log(\hat{y}) + (1 - y) \log(1 - \hat{y}))
$$

The logic of this function becomes clear when we look at the two cases of the true label $y$:
- **Case y = 1:** The loss function becomes $L = -\log(\hat{y})$. For this loss function to reach its minimum value, $\log(\hat{y})$ must be as large as possible. Since $\hat{y}$ is bounded by the Sigmoid function (it cannot exceed 1), the algorithm will try to adjust the parameters so that $\hat{y}$ gets as close to 1 as possible.
- **Case y = 0:** The loss function becomes $L = -\log(1 - \hat{y})$. To minimize $L$, the value $\log(1 - \hat{y})$ must be as large as possible, which means $\hat{y}$ must get as close to 0 as possible.

Thus, in both cases, this loss function highly effectively forces the predicted value $\hat{y}$ to move in the correct direction of the true label.

### 3. Cost Function for the Entire Dataset
While the Loss Function only evaluates a single example, the Cost Function, denoted as $J(w, b)$, measures the model's performance over the entire training set.

The cost function is simply the average of all the Loss Functions from the $1$-st example to the $m$-th example:
$$
J(w, b) = \frac{1}{m} \sum_{i=1}^m L(\hat{y}^{(i)}, y^{(i)})
$$

When fully expanded, the formula for the Cost Function in Logistic Regression is:
$$
J(w, b) = -\frac{1}{m} \sum_{i=1}^m [y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)})]
$$

**Summary of the training process:** The essence of training a Logistic Regression model (or a single neuron) is the process of searching for the parameter set $w$ and $b$ such that the value of the Cost Function $J(w, b)$ reaches its minimum. This minimization process will be executed through the Gradient Descent algorithm in the upcoming lessons.

## Gradient Descent
### 1. The Core Idea of Gradient Descent
The goal of the Logistic Regression algorithm is to find the parameter set $w$ (weight) and $b$ (bias) such that the Cost Function $J(w, b)$ reaches its minimum possible value.

Imagine the Cost Function $J(w, b)$ as a three-dimensional surface. The vertical axis represents the cost function's value, while the horizontal plane below represents the space of the parameters $w$ and $b$.

Fortunately, as explained in the previous section, thanks to the use of the Log-Loss function, the Cost Function $J(w, b)$ in Logistic Regression is a strictly convex function. Its shape resembles a large, smooth bowl with only one single bottom point, which is the global optimum. It is not bumpy and does not have multiple false bottoms, known as local optima, like non-convex functions.


The Gradient Descent algorithm operates on the following principle:
- **Initialization:** Start at any random point on the bowl's surface, typically initializing $w$ and $b$ to 0.
- **Measuring the slope:** At the current position, the algorithm calculates the steepest downward direction.
- **Descending:** Take a step in that downward direction.
- **Repeating:** This process repeats continuously until the algorithm converges at the lowest point of the bowl, which is the point of minimum cost.

<img src="../images/Week02/image7.png" width=800>

### 2. Parameter Update Rule
To execute the descent step, the algorithm uses the following mathematical update formula for each iteration:
$$
w := w - \alpha \frac{\partial J(w, b)}{\partial w}
$$
$$
b := b - \alpha \frac{\partial J(w, b)}{\partial b}
$$
There are two key components in this formula:
- **Derivative (Gradient):** The term $\frac{\partial J(w, b)}{\partial w}$ is the slope of the function at the current point. It indicates whether the function is sloping upward or downward along the $w$ axis, and how steep that slope is. The derivative acts as a directional guide pointing toward the lowest point.
  - If the current point is on the right slope where the derivative is positive, the algorithm will subtract a positive number, causing $w$ to decrease and shifting the point leftward toward the bottom.
  - If the current point is on the left slope where the derivative is negative, the algorithm will subtract a negative number, effectively adding to it, causing $w$ to increase and shifting the point rightward toward the bottom.
- **Learning Rate ($\alpha$):** The parameter $\alpha$ controls the size of each step. If $\alpha$ is too large, the step will be too wide, potentially overshooting the minimum point. If $\alpha$ is too small, the model will inch forward gradually, taking a massive amount of time to converge.

<img src="../images/Week02/image8.png" width=800>

### 3. Mathematical and Programming Notation Conventions
Andrew Ng particularly emphasizes how to translate mathematical language into programming language to avoid confusing those who are not familiar with calculus.

**Mathematically:**
- The lowercase letter $d$, as in $\frac{dJ}{dw}$, is used for standard derivatives when the function $J$ depends on only one single variable.
- The curly $\partial$ symbol, as in $\frac{\partial J}{\partial w}$, is used to denote a partial derivative. It is used when the function $J$ depends on two or more variables, which are $w$ and $b$ in this case. Its practical meaning is that we are measuring the slope of the function $J$ with respect to the variable $w$ while temporarily treating the variable $b$ as a constant.

**Programmatically:**

When writing code in Python, we do not use complex mathematical symbols. To represent derivative terms, the standard variable naming convention in this course is to add the prefix $d$ before the variable name:
- The variable dw will represent the value of the partial derivative term $\frac{\partial J(w, b)}{\partial w}$.
- The variable db will represent the value of the partial derivative term $\frac{\partial J(w, b)}{\partial b}$.

Thus, the parameter update code will look incredibly simple:
```python
w = w - alpha * dw
b = b - alpha * db
```

## Derivatives with a Computation Graph
### 1. The Core Idea of Computation Graphs and Backpropagation
A computation graph breaks down a complex equation into the most basic computational steps (nodes). In this example, the function $J = 3(a + bc)$ is split into 3 steps:
- **Step 1:** Calculate $u = bc$
- **Step 2:** Calculate $v = a + u$
- **Step 3:** Calculate $J = 3v$

With the input values $a=5, b=3, c=2$, we compute forward (forward propagation) from left to right to get:
- $u = 3 \times 2 = 6$
- $v = 5 + 6 = 11$
- $J = 3 \times 11 = 33$

<img src="../images/Week02/image9.png" width=800>

The goal of the Backpropagation algorithm is to find out: If we nudge the initial input variables ($a, b, c$) up a little bit, how will the final output value $J$ change? In other words, finding the derivatives $\frac{\partial J}{\partial a}$, $\frac{\partial J}{\partial b}$, and $\frac{\partial J}{\partial c}$.

Unlike forward propagation which goes from left to right, backpropagation will go from right to left (from the output variable $J$ moving gradually backward to the input variables), leveraging the newly computed intermediate derivatives to calculate the subsequent ones. This is exactly how computers save a tremendous amount of computational resources.

### 2. Variable Naming Convention in Python
Before getting into the calculations, Andrew Ng introduces a core convention used throughout the course to make writing Python code extremely concise.

Instead of writing lengthy variable names like **dJ_dvar** to represent the derivative of the cost function $J$ with respect to a certain variable var ($\frac{\partial J}{\partial var}$), we simply name that variable **dvar**.
- For example: The derivative $\frac{\partial J}{\partial v}$ will be written concisely in code as dv.
- The derivative $\frac{\partial J}{\partial a}$ will be written concisely as da.
- Similarly, we have du, db, dc.

All variables starting with the letter d in the code implicitly mean the **derivative of the final output variable (here being J) with respect to that variable**.

### 3. Step-by-Step Execution of Backpropagation
**Step 1: Calculate the derivative at the Final Node (Finding dv)**
- Equation: $J = 3v$
- Intuition: If $v$ increases by $1$ unit, $J$ will increase by $3$ times.
- Derivative: $\frac{\partial J}{\partial v} = 3$. In code, we write dv = 3.

**Step 2: Move back one Node (Finding da and du)**
- Equation: $v = a + u$
- To find how $a$ affects $J$, we use the Chain Rule in Calculus. The effect of $a$ on $J$ will equal: The degree to which $a$ changes $v$, multiplied by the degree to which $v$ changes $J$.
- Mathematics: $\frac{\partial J}{\partial a} = \frac{\partial J}{\partial v} \times \frac{\partial v}{\partial a}$
- Since $v = a + u$, if $a$ increases by $1$ unit, $v$ also increases by $1$ unit. So $\frac{\partial v}{\partial a} = 1$.
- Plugging in the numbers: $\frac{\partial J}{\partial a} = 3 \times 1 = 3$. In code, we write da = 3.

Similarly for variable $u$:
- Mathematics: $\frac{\partial J}{\partial u} = \frac{\partial J}{\partial v} \times \frac{\partial v}{\partial u}$
- Since $v = a + u$, the change of $v$ with respect to $u$ is $1$ ($\frac{\partial v}{\partial u} = 1$).
- Plugging in the numbers: $\frac{\partial J}{\partial u} = 3 \times 1 = 3$. In code, we write du = 3.

**Step 3: Move back to the First Node (Finding db and dc)**

Now we continue to find the effect of $b$ on $J$. This effect must pass through the intermediate variable $u$.
- Equation: $u = bc$ (with $c=2$)
- Chain rule: $\frac{\partial J}{\partial b} = \frac{\partial J}{\partial u} \times \frac{\partial u}{\partial b}$
- Note that we have already calculated $\frac{\partial J}{\partial u} = 3$ (which is the variable du in Step 2). This is the power of the computation graph: reusing calculated results!
- Calculating the rest: From $u = b \times 2$, we see that if $b$ increases by $1$, $u$ doubles. So $\frac{\partial u}{\partial b} = 2$.
- Plugging in the numbers: $\frac{\partial J}{\partial b} = 3 \times 2 = 6$. In code, we write db = 6.

Similarly for variable $c$:
- Chain rule: $\frac{\partial J}{\partial c} = \frac{\partial J}{\partial u} \times \frac{\partial u}{\partial c}$
- Calculating the intermediate derivative: From $u = 3 \times c$ (since $b=3$), if $c$ increases by $1$, $u$ increases by $3$ times. So $\frac{\partial u}{\partial c} = 3$.
- Plugging in the numbers: $\frac{\partial J}{\partial c} = 3 \times 3 = 9$. In code, we write dc = 9.

<img src="../images/Week02/image10.png" width=800>

**Conclusion:**
Through the computation graph, calculating derivatives (Backpropagation) becomes highly systematic and resource-efficient. The computer only needs to go backward from right to left, multiplying local gradients with the derivative passed from the node behind it.

## Logistic Regression Gradient Descent
### 1. Setting up the Computation Graph for Logistic Regression
For easy visualization, let's consider an extremely simple Logistic Regression example: The input has only 2 features, $x_1$ and $x_2$. Accordingly, the model will have 2 weights, $w_1$ and $w_2$, and a bias, $b$.

The Forward Propagation process to calculate the predicted value and the error is divided into 3 steps (3 nodes) on the graph:
- **Step 1 - Calculate the linear value ($z$):**
  $$
  z = w_1x_1 + w_2x_2 + b
  $$
- **Step 2 - Calculate the predicted value ($a$ or $\hat{y}$):** Apply the Sigmoid activation function to $z$.
  $$
  a = \sigma(z) = \frac{1}{1 + e^{-z}}
  $$
- **Step 3 - Calculate the error (Loss Function $L$):** Use the Log-Loss function to measure the discrepancy between the prediction $a$ and the true label $y$.
  $$
  L(a, y) = -(y \log(a) + (1 - y) \log(1 - a))
  $$

The goal of the Gradient Descent algorithm is to find the derivative of the Loss Function $L$ with respect to the parameters $w_1, w_2, b$ to know how to adjust them. We will do this by Backpropagating from right to left on the graph.

<img src="../images/Week02/image11.png" width=800>

### 2. The Backpropagation Process (Calculating Derivatives)
We will move backward through each node, applying the Chain Rule and using the variable naming convention (adding the prefix d) learned in the previous lesson.

**Step 1: Calculate the derivative of $L$ with respect to $a$ (Denoted as da)**
- **Goal:** Calculate $\frac{\partial L}{\partial a}$.
- **Result:** By taking the direct derivative of the Log-Loss function, we get the formula:
  $$
  da = -\frac{y}{a} + \frac{1 - y}{1 - a}
  $$

**Step 2: Calculate the derivative of $L$ with respect to $z$ (Denoted as dz)**
- **Goal:** Find out how $z$ affects $L$. Intuition shows that this effect must pass through the intermediate variable $a$.
- **Chain Rule:** $\frac{\partial L}{\partial z} = \frac{\partial L}{\partial a} \times \frac{\partial a}{\partial z}$
  - The first component $\frac{\partial L}{\partial a}$ is exactly da that we just calculated in Step 1.
  - The second component $\frac{\partial a}{\partial z}$ is the derivative of the Sigmoid function. Fortunately, the derivative of the Sigmoid function has a highly elegant mathematical property: $\frac{\partial a}{\partial z} = a(1 - a)$.
- **Result:** When you multiply da by $a(1 - a)$ and simplify the algebraic expressions, everything cancels each other out perfectly, leaving an extremely minimalist result:
  $$
  dz = a - y
  $$

**Step 3: Calculate the derivative of $L$ with respect to the parameters $w_1, w_2, b$**

Now we continue backward to the first node. From the equation $z = w_1x_1 + w_2x_2 + b$, combined with the newly found dz, we calculate the final derivatives:
- **Calculate dw1 ($\frac{\partial L}{\partial w_1}$):** According to the chain rule, $\frac{\partial L}{\partial w_1} = \frac{\partial L}{\partial z} \times \frac{\partial z}{\partial w_1}$. Since $\frac{\partial z}{\partial w_1} = x_1$, we have:
  $$
  dw_1 = x_1 \times dz
  $$
- **Calculate dw2 ($\frac{\partial L}{\partial w_2}$):** Similarly:
  $$
  dw_2 = x_2 \times dz
  $$
- **Calculate db ($\frac{\partial L}{\partial b}$):** Since $\frac{\partial z}{\partial b} = 1$, we have:
  $$
  db = dz
  $$

### 3. Parameter Update

Once we have all the derivatives dw1, dw2, and db calculated, the final action is to take a step to adjust these parameters, with $\alpha$ being the Learning Rate:
$$
w_1 := w_1 - \alpha \times dw_1
$$
$$
w_2 := w_2 - \alpha \times dw_2
$$
$$
b := b - \alpha \times db
$$

<img src="../images/Week02/image12.png" width=800>

**Summary:** The mathematical equations above are the core of a single-neuron neural network. You compute forward to find the prediction, then use that exact prediction to compute backward to find the error, and finally use that error to update the parameters. Everything happens in a highly systematic and rhythmic manner.

## Gradient Descent on m Examples
### 1. The Nature of Derivatives across the Entire Training Set
As we know, the Cost Function $J(w, b)$ is the average of the Loss Functions $L$ across all $m$ training examples.
$$
J(w, b) = \frac{1}{m} \sum_{i=1}^{m} L(a^{(i)}, y^{(i)})
$$
From this definition, mathematics dictates that: The derivative of the overall Cost Function $J$ with respect to a parameter (e.g., $w_1$) is also the **average of the derivatives** calculated on each individual training example with respect to that parameter.

Our task is to iterate through each data sample, calculate the local derivatives (such as $dw_1^{(i)}$, $dw_2^{(i)}$, $db^{(i)}$), accumulate them, and then divide by $m$ to get the overall derivatives ($dw_1, dw_2, db$). Finally, we use these overall derivatives to update the parameters.

<img src="../images/Week02/image13.png" width=800>

### 2. Gradient Descent Algorithm Using Loops
Assuming our input vector has only 2 features ($n = 2$, meaning only $x_1$ and $x_2$). The process of executing one step (one iteration) of Gradient Descent occurs in the following 4 stages:

**Stage 1: Initialize accumulators**

We need variables to store the sum of values during the loop. Initially, all are set to 0.
- $J = 0$
- $dw_1 = 0$
- $dw_2 = 0$
- $db = 0$

**Stage 2: Loop through $m$ training examples (from $i = 1$ to $m$)**

For each data sample $i$:
- **Forward Propagation:**

  Calculate the linear value: $z^{(i)} = w_1x_1^{(i)} + w_2x_2^{(i)} + b$

  Calculate the prediction: $a^{(i)} = \sigma(z^{(i)})$

  Accumulate the cost: $J \mathrel{+}= -[y^{(i)} \log(a^{(i)}) + (1 - y^{(i)}) \log(1 - a^{(i)})]$

- **Backward Propagation:**

  Calculate the error: $dz^{(i)} = a^{(i)} - y^{(i)}$

  Accumulate the derivatives:

  $dw_1 \mathrel{+}= x_1^{(i)} dz^{(i)}$

  $dw_2 \mathrel{+}= x_2^{(i)} dz^{(i)}$

  $db \mathrel{+}= dz^{(i)}$

**Stage 3: Calculate the average**

After the loop ends, we divide the accumulated variables by $m$ to get the actual derivatives and cost.
- $J \mathrel{/}= m$
- $dw_1 \mathrel{/}= m$
- $dw_2 \mathrel{/}= m$
- $db \mathrel{/}= m$

**Stage 4: Update parameters**

Use the newly calculated overall derivatives to take a step in the direction of steepest descent, with $\alpha$ being the learning rate.
- $w_1 := w_1 - \alpha \times dw_1$
- $w_2 := w_2 - \alpha \times dw_2$
- $b := b - \alpha \times db$

The entire 4 stages above only complete one single step of the algorithm. In practice, you will have to wrap this entire block of code in another massive loop (running thousands of times) until the model converges.

<img src="../images/Week02/image14.png" width=800>

### 3. The Loop Bottleneck and the Need for Vectorization

Although the above algorithm is perfectly correct mathematically, it is extremely inefficient when implemented in actual code. The reason is that it contains explicit for loops.

If you look closely, you will see that we are facing two nested loops:
- The outer loop iterating through $m$ training examples (the number $m$ can be up to millions of images).
- The inner loop iterating through $n$ features to update from $dw_1, dw_2$ up to $dw_n$ (if the image is 64x64 pixels, then $n = 12,288$). You cannot manually write code for every single line $dw_1, dw_2...$ and must use a loop for this part.

In the Deep Learning era, running such nested loops in Python will make the processing speed unacceptably slow. Your algorithm might take days instead of minutes to finish running.

The solution to this problem is the **Vectorization** technique. This technique leverages the parallel architecture of hardware (CPU/GPU) to perform millions of calculations simultaneously using matrix multiplication, helping to completely eliminate cumbersome loops in the source code.
