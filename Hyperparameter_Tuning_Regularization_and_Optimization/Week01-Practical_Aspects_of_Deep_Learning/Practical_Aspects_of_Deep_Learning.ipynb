# Data Splitting: Training Set, Validation Set, and Test Set

In the process of building a Machine Learning or Deep Learning system, establishing a precise methodology to evaluate model performance is a crucial factor. This process begins by dividing the initial dataset into three distinct subsets.

## 1. Function of Each Data Subset

The strict separation among these three datasets ensures objectivity in evaluating the model's generalization capacity on unseen data.

*   **Training Set:** This is the core foundation for the model to build knowledge representations. The optimization algorithm will access data from this set repeatedly to update the neural network's internal parameters (weight matrix $W$ and bias vector $b$) in order to minimize the error.
*   **Validation Set / Dev Set:** This dataset functions as an intermediate measurement system. Based on the performance metrics on the validation set, engineers will fine-tune the hyperparameters, such as the learning rate $\alpha$, network architecture, or the degree of regularization.
*   **Test Set:** This is the final and most independent evaluation mechanism. It is used exclusively once at the final stage to provide an objective estimate of the system's actual performance. It is strictly forbidden to use results from the test set to retrospectively alter the model, in order to prevent data leakage.

## 2. Data Allocation Ratios Based on Project Scale

The methodology for proportioning datasets has shifted significantly alongside the evolution from traditional data scales to big data.

### The Era of Traditional Machine Learning
When the total data volume is relatively modest (from a few hundred to approximately ten thousand samples), the classical splitting ratio is typically 60/20/20 for the Training Set, Validation Set, and Test Set, respectively, or 70/30 if the Test Set is omitted.

### The Era of Modern Deep Learning
With the explosion of big data, models can be trained on millions of samples. A validation set containing approximately $10,000$ samples (equivalent to $1\%$ or less) generally provides sufficient statistical significance. Therefore, the current trend prioritizes maximum resources for training, with prevalent allocation ratios such as 98/1/1, or 99.5/0.25/0.25 for extremely large datasets.

## 3. Distribution Matching

Determining an appropriate splitting ratio must be accompanied by adherence to a fundamental statistical principle: The Validation Set and the Test Set must be extracted from the same data distribution, and this distribution must accurately reflect the real-world problem.

A mismatch in data distribution is a common cause of failure for Deep Learning projects upon practical deployment. The development team must construct the Validation Set and the Test Set purely based on the actual target data, aiming to direct the entire development process toward the core problem. Auxiliary data volumes or data collected from other sources should be entirely consolidated into the Training Set.

# Bias and Variance in Deep Learning

In the domain of Deep Learning, the analysis of Bias and Variance serves as the core methodology for diagnosing performance and directing the optimization of neural network models. This process enables engineers to precisely identify the root causes of sub-optimal system behavior before applying any adjustment techniques.

## 1. Definitions of Bias and Variance



These two metrics represent the primary sources of error during data modeling.

*   **Bias:** This metric quantifies the deviation between the model's average prediction and the actual value within the dataset. A model exhibiting High Bias typically demonstrates Underfitting. This state occurs when the neural network architecture is too simplistic or the optimization process is insufficiently deep for the system to capture the complex internal structures and patterns of the training data. For instance, in the development of a specialized system like a Vietnamese legal question-answering platform (VNLaw-QA), high bias might manifest if the underlying model architecture is too shallow to comprehend the intricate syntactical dependencies and semantic nuances inherent in legal text.
  
*   **Variance:** This metric measures the fluctuation in the model's predictions when trained on varying datasets. A model with High Variance leads to Overfitting. In this scenario, the neural network possesses excessive complexity, resulting in the memorization of random noise within the training set. Consequently, the model loses its generalization capacity and produces inaccurate predictions when confronted with entirely new data. In the context of a legal Q&A system, this would equate to the model perfectly reciting answers for the specific phrasing present in the training corpus, but failing entirely when a user inputs a newly formulated, unseen query.

## 2. Diagnostic Methodology via Error Analysis

To accurately diagnose the state of a neural network, an engineer must establish a benchmark known as Optimal Error or Bayes Error. In practice, Human-level Performance is frequently utilized as a proxy for this benchmark. The diagnostic process relies on evaluating the discrepancies among three indicators: Bayes Error, Training Error, and Validation Error.

Assuming the Bayes Error for a specific classification task approaches $1\%$, the analytical scenarios are defined as follows:

*   **High Bias:** Training Error is $15\%$, Validation Error is $16\%$. The significant gap between the training error and the Bayes benchmark indicates that the neural network cannot adequately solve the problem even on the source dataset, which is a clear manifestation of a Bias issue.
*   **High Variance:** Training Error is $1\%$, Validation Error is $11\%$. The model achieves near-perfect accuracy during training, but performance degrades severely on unseen data. The large discrepancy between the training error and validation error is the hallmark of high Variance.
*   **High Bias and High Variance:** Training Error is $15\%$, Validation Error is $30\%$. This represents the most inefficient state, where the model simultaneously fails to extract fundamental features and overfits locally to anomalies within the dataset.
*   **Optimal State:** Training Error is $1\%$, Validation Error is $2\%$. The model achieves an ideal equilibrium, possessing both robust learning capacity and high generalization capability.

## 3. The Deep Learning Paradigm Shift

In traditional Machine Learning theory, engineers consistently faced the Bias-Variance Tradeoff principle. According to this paradigm, altering the model structure to minimize one metric would inevitably cause an increase in the other.

However, the emergence of modern Deep Learning architectures has introduced a distinct approach. By leveraging the capability to train massive neural networks on exceptionally large datasets, coupled with advanced optimization algorithms, contemporary engineers can deploy independent interventions to simultaneously mitigate both Bias and Variance without strictly enduring the compromises required by classical statistical models.

## 4. Remediation and Optimization Strategies

The standardized protocol for neural network development requires engineering teams to sequentially apply remediation strategies based on the diagnostic results:

*   **Mitigating High Bias:** The primary objective is to enhance the representational capacity of the model. Implementation techniques include increasing the number of hidden units, adding more layers to deepen the network, extending the training duration (increasing epochs), or migrating to advanced network architectures that better align with the structural properties of the data.
*   **Mitigating High Variance:** Once the network is sufficiently complex to achieve a low training error but lacks generalization, the most definitive solution is to procure a larger and more diverse volume of training data. Under constraints where data resources are limited, the application of Regularization methods becomes mandatory to constrain and restrict the excessive growth of weight parameters. These techniques include $L_1$ and $L_2$ regularization, Dropout (the stochastic deactivation of neurons), Data Augmentation, or Early Stopping.

# The Basic Recipe for Machine Learning

When training an artificial neural network, a systematic methodology is required to diagnose and resolve performance issues. This process, often referred to as the basic recipe for machine learning, relies on analyzing training and development errors to identify high bias or high variance.

## 1. Diagnosing and Mitigating High Bias

After training an initial model, the first diagnostic step is to evaluate its performance on the training dataset. If the algorithm exhibits high bias, it will fail to fit the training data accurately. To mitigate high bias, the following strategies should be applied iteratively:

*   **Increase Network Capacity:** Expand the model by adding more hidden layers or increasing the number of hidden units. Utilizing a larger neural network is consistently beneficial for reducing bias.
*   **Extend Training Duration:** Training the algorithm for a longer period or utilizing more advanced optimization algorithms can facilitate better convergence. While training longer does not guarantee improvement, it is never detrimental.
*   **Explore Alternative Architectures:** Searching for a new neural network architecture that is more inherently suited to the specific problem can occasionally resolve bias issues.

This phase must be repeated until the model successfully escapes high bias and fits the training data proficiently, assuming the fundamental baseline error is reasonably achievable.

## 2. Diagnosing and Mitigating High Variance

Once the bias is reduced to an acceptable threshold, the subsequent step is to evaluate the algorithm's performance on the development set to determine if it suffers from high variance. High variance indicates an inability to generalize from the training data to the development data. If high variance is detected, the following interventions are recommended:

*   **Acquire Additional Data:** Providing the model with more training data is the most effective approach to resolving high variance.
*   **Apply Regularization:** In scenarios where acquiring more data is unfeasible, implementing regularization techniques is crucial to reduce overfitting.
*   **Explore Alternative Architectures:** Similar to addressing bias, discovering a more appropriate network architecture can sometimes reduce variance.

The diagnostic loop continues, alternating between bias and variance checks, until the model achieves an optimal state characterized by both low bias and low variance.

## 3. The Deep Learning Paradigm Shift

A critical realization in this diagnostic process is that the specific problem—high bias versus high variance—dictates entirely different subsets of solutions. For instance, acquiring more training data is completely ineffective if the underlying issue is high bias.

In the preceding era of traditional machine learning, practitioners continuously navigated the bias-variance tradeoff. Due to a lack of independent optimization tools, interventions that decreased bias frequently increased variance, and conversely, decreasing variance often compromised bias.

However, the modern era of big data and deep learning has fundamentally altered this paradigm. Assuming the availability of extensive data and computational resources, engineers now possess decoupled mechanisms to optimize performance:

*   **Scaling the Network:** Training a larger neural network almost exclusively reduces bias without significantly inflating variance, provided that appropriate regularization is applied. The primary cost associated with excessively large networks is merely computational time.
*   **Scaling the Data:** Acquiring more data exclusively reduces variance without negatively impacting the bias.

While regularization—a primary tool for variance reduction—may marginally increase bias, this side effect is typically negligible when the neural network is sufficiently large. Consequently, modern deep learning provides a broader array of options to independently minimize bias or variance without being strictly constrained by historical tradeoffs.
